# Task-B

```python
path = "../data/faina-v1.0_fadeit-shared-task/faina-v1.0_fadeit-shared-task/data/subtask-b/train-dev.conll"
```

```python
def show_conll_samples(path, n=2):
    samples = []
    current = []

    with open(path, encoding="utf-8") as f:
        for line in f:
            if line.strip() == "":
                if current:
                    samples.append("".join(current))
                    current = []
                continue

            current.append(line)

        # Append last block
        if current:
            samples.append("".join(current))

    # Print the first n samples
    for i, sample in enumerate(samples[:n], 1):
        print(f"\n===== SAMPLE {i} =====\n")
        print(sample)
```

## Raw samples

```python
show_conll_samples(path, n=2)
```

```python
def read_conll(path):
    rows = []
    current_post_id = None

    with open(path, encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line.startswith("# post_id"):
                current_post_id = line.split("=")[1].strip()
                continue
            if not line or line.startswith("#"):
                continue
            token_id, token, label1, label2 = line.split("\t")

            rows.append({
                "post_id": current_post_id,
                "token_id": int(token_id),
                "token": token,
                "labels_a1": label1,
                "labels_a2": label2,
            })

    return pd.DataFrame(rows)
```

```python
df_B = read_conll(path)
df_B.shape
```

## Label distribution

```python
all_labels = (df_B['labels_a1'].dropna().astype(str).str.split('|').explode().str.strip())
all_labels = all_labels[all_labels != "O"]
label_counts = all_labels.value_counts().reset_index()
label_counts.columns = ["label", "count"]
print(label_counts)
```

```python
chart = (
    alt.Chart(label_counts)
    .mark_bar()
    .encode(
        x=alt.X("label:N", sort='-y', title="labels_a1"),
        y=alt.Y("count:Q", title="Count"),
        tooltip=["label", "count"],
    )
    .properties(
        title="Distribution of labels_a1 (excluding O)",
        width=700,
        height=400,
    )
)

chart
```
